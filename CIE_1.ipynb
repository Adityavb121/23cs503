```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/mtcars.csv')
transmission_counts = df['am'].value_counts()
transmission_counts.plot(kind='bar')
plt.title('Frequency Distribution of Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Frequency')
plt.show()
```

```python
df = pd.read_csv('/content/mtcars.csv')
plt.boxplot(df['mpg'])
plt.title("Box Plot of MPG")
plt.ylabel("Miles per Gallon")
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv('/content/mtcars.csv')
auto = (df['am'] == 0).sum()
manual = (df['am'] == 1).sum()
plt.bar(['Automatic', 'Manual'], [auto, manual], color=['orange', 'blue'])
plt.title("Frequency of Transmission Types")
plt.xlabel("Type of cars")
plt.ylabel("Number of Cars")
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv('/content/mtcars.csv')

plt.scatter(df['wt'], df['mpg'], color='green')
plt.title("Car Weight vs MPG")
plt.xlabel("Weight")
plt.ylabel("Miles per Gallon")
plt.grid(True)
plt.show()
```
