```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('titanic')

# Select numerical columns
num_cols = df[['age', 'fare', 'pclass']].dropna()

# Covariance matrix
cov_matrix = num_cols.cov()
print("Covariance:\n", cov_matrix)

# Correlation matrix
corr_matrix = num_cols.corr()
print("\nCorrelation:\n", corr_matrix)
```

Visualizing Correlation

```python
# Heatmap of correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Step 0: Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Step 1: Load the Titanic dataset
df = pd.read_csv('/Titanic-Dataset.csv')

# Step 2: Exploratory Data Analysis (EDA) & Central Tendency
print("\n--- Central Tendency (Age & Fare) ---")
print("Age - Mean:", df['Age'].mean())
print("Age - Median:", df['Age'].median())
print("Age - Mode:", df['Age'].mode()[0])

print("Fare - Mean:", df['Fare'].mean())
print("Fare - Median:", df['Fare'].median())
print("Fare - Mode:", df['Fare'].mode()[0])

# Step 3: Replace 0/1 in 'Survived' column with labels and plot bar chart
df['Survived'] = df['Survived'].replace({0: 'Not Survived', 1: 'Survived'})

# Plot bar chart for Survived and Pclass
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")

plt.subplot(1, 2, 2)
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Count")

plt.tight_layout()
plt.show()

# Step 4: Find Null Values and Fill with Mean
print("\n--- Null Value Count ---")
print(df.isnull().sum())

# Fill nulls in Age and Fare with their means
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)

# Verify if nulls are filled
print("\n--- Nulls after Filling ---")
print(df[['Age', 'Fare']].isnull().sum())

# Step 5: Detect Outliers using Z-score
z_scores = zscore(df[['Age', 'Fare']])
abs_z_scores = np.abs(z_scores)

# Outlier detection threshold
threshold = 3
outliers = (abs_z_scores > threshold)

# Print number of outliers
print("\n--- Outlier Detection ---")
print("Age Outliers:", np.sum(outliers[:, 0]))
print("Fare Outliers:", np.sum(outliers[:, 1]))

# Step 6: Spot the Outliers using Boxplot
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=df['Age'])
plt.title("Boxplot for Age")

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Fare'])
plt.title("Boxplot for Fare")

plt.tight_layout()
plt.show()
```

```python
# Load the Titanic dataset. Import the libraries.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore

df = pd.read_csv('/Titanic-Dataset.csv')
```

```python
# Perform the Exploratory Data Analysis and display the Central Tendency for Age and Fare.

# Central Tendency for Age
print("Age - Mean:", df['Age'].mean())
print("Age - Median:", df['Age'].median())
print("Age - Mode:", df['Age'].mode()[0])

# Central Tendency for Fare
print("Fare - Mean:", df['Fare'].mean())
print("Fare - Median:", df['Fare'].median())
print("Fare - Mode:", df['Fare'].mode()[0])
```

```python
#Q3) Replace the values of Survived column from 1 and 0 to 'Survived' and 'Not Survived'. Plot bar charts for Survived and Pclass.

df['Survived'] = df['Survived'].map({0: 'Not Survived', 1: 'Survived'})
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
print("Null values before filling:\n", df.isnull().sum())

df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)

print("Null values after filling:\n", df[['Age', 'Fare']].isnull().sum())
```

```python
z_scores = zscore(df[['Age', 'Fare']])
outliers = (np.abs(z_scores) > 3)

print("Number of Age Outliers:", np.sum(outliers[:, 0]))
print("Number of Fare Outliers:", np.sum(outliers[:, 1]))
```
